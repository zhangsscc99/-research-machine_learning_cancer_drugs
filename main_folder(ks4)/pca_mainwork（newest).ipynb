```python
#do some further processing with "A" because there are some data error in it
def remove_extradots(string):
    counter=0
    lst=[]
    for i in range(len(string)):
        if string[i]=='.' and counter<1 :
            lst.append(string[i])
            counter+=1
        elif string[i]!='.':
            lst.append(string[i])
    res="".join(lst)
    return res



import csv
res_list=[]
print(remove_extradots("-0.29444.2"))
with open('A.csv') as file2:
    reader=csv.reader(file2)

    for row in reader:
        data=[]
        for i in range(len(row)):
            
            data.append(remove_extradots(row[i]))
        res_list.append(data)
    


with open('modified_A.csv', 'w') as file3:
    writer = csv.writer(file3)
    writer.writerows(res_list)
file3.close()
                
```

```python
#a code block that is used to deal with the data using PCA models(dimensionality reduction)
import pandas as pd
import plotly.express as px
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X_drug = e_and_p.iloc[:, 3:240].values

X_drug = StandardScaler().fit_transform(X_drug)

pca_drug = PCA(n_components=6)
components = pca_drug.fit_transform(X_drug)
principalDf = pd.DataFrame(data = components
             , columns =['principal component 1', 'principal component 2','principal component 3','principal component 4','principal component 5','principal component 6'])
principalDf
```

```python
#wipe out those whose concentration is not valid
def get_centroid(point_lst):
    x_sum=0
    y_sum=0
    z1_sum=0
    z2_sum=0
    z3_sum=0
    z4_sum=0


    for i in range(len(point_lst)):
        x_sum+=point_lst[i][0]
        y_sum+=point_lst[i][1]
        z1_sum+=point_lst[i][2]
        z2_sum+=point_lst[i][3]
        z3_sum+=point_lst[i][4]
        z4_sum+=point_lst[i][5]


    x_mean=x_sum/len(point_lst)
    y_mean=y_sum/len(point_lst)
    z1_mean=z1_sum/len(point_lst)
    z2_mean=z2_sum/len(point_lst)
    z3_mean=z3_sum/len(point_lst)
    z4_mean=z4_sum/len(point_lst)
    centroid=[]
    centroid.append(x_mean)
    centroid.append(y_mean)
    centroid.append(z1_mean)
    centroid.append(z2_mean)
    centroid.append(z3_mean)
    centroid.append(z4_mean)
    centroid.append(point_lst[0][-1])
    return centroid

```

```python
#calculate DMSO circle
p_DMSO=[]
for i in range(24,72):
    p_DMSO.append([principalDf.iloc[i,:][0],principalDf.iloc[i,:][1],principalDf.iloc[i,:][2],principalDf.iloc[i,:][3],principalDf.iloc[i,:][4],principalDf.iloc[i,:][5],'DMSO+DMSO'])
p_DMSO[0]
```

```python
#get the centroid of DMSO
centroid_DMSO=get_centroid(p_DMSO)
centroid_DMSO
```

```python
#get the DMSO circle
#
#
#
#max_dist we got
from math import sqrt
def euclidean_distance(row1, row2):
    distance =0
	#print(len(row1)-1) this is to ensure that we calculate 6 points
    for i in range(len(row1)-1):

	

        distance += (row1[i] - row2[i])**2
    return sqrt(distance)

	

import math
dist_lst=[]
for i in range(len(p_DMSO)):

    dist_lst.append(euclidean_distance(centroid_DMSO,p_DMSO[i]))
#print(dist_lst)
print(len(dist_lst))
dist_lst.sort()

max_dist=dist_lst[-1]
print(max_dist)
```

```python
#still wiping out not effective concentration
p=[]
for i in range(24):
    p.append([principalDf.iloc[i,:][0],principalDf.iloc[i,:][1],principalDf.iloc[i,:][2],principalDf.iloc[i,:][3],principalDf.iloc[i,:][4],principalDf.iloc[i,:][5]])
for i in range(6):
    p[i].append('Etoposide')
for j in range(6):
    p[j+6].append('Paclitaxel')
for k in range(6):
    p[k+12].append('Etoposide+Paclitaxel')
for m in range(6):
    p[m+18].append("Paclitaxel+Etoposide")
#print(p)
dist_lst2=[]
for i in range(24):
    dist_lst2.append(euclidean_distance(centroid_DMSO,p[i]))
#print(dist_lst2)
#record the index
#then, according to the index, wipe out those same index in p
effect=[]
indexlst=[]
for i in range(12):
    dist=euclidean_distance(centroid_DMSO,p[i])
    
    if dist>max_dist:
        effect.append(p[i])
        indexlst.append(i)
#24 becomes 24
print(effect,len(effect),indexlst)
#1,2,3
#1,2,3,4
```

```python
#again. computing distances.
#0-5  etoposide
#6-11 paclitaxel
#plus 12 first. then plus 18
import math
distance=[]
#between effective etoposide and combination (average distance)
for i in range(3):
    distance.append(euclidean_distance(p[indexlst[i]],p[i+12]))
distance=sum(distance)/len(distance)
print(distance)
distance2=[]
for j in range(3,7):
    distance2.append(euclidean_distance(p[indexlst[j]],p[j+6]))
distance2=sum(distance2)/len(distance2)
print(distance2)
#since distance<distance2, etoposide is more similar to the combination
```

```python
#the last set of the combination
distance3=[]
for m in range(3):
    distance3.append(euclidean_distance(p[indexlst[m]],p[m+18]))
print(distance3)
distance3=sum(distance3)/len(distance3)

distance4=[]
for n in range(3,7):
    distance4.append(euclidean_distance(p[indexlst[n]],p[n+12]))
distance4=sum(distance4)/len(distance4)
print(distance3,distance4)
#still etoposide is more effective. 
#one question 
#this time, conclusion: etoposide is more effective
```
