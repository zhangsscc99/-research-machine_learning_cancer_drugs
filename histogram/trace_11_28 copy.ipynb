```python
def merge(start,end,i):
    import os
    list_image=[i for i in os.listdir('/Users/mac/Desktop/pca_knn_cancer_drug/pca_knn/histogram') if i.endswith(".jpeg")]
    list_image.sort()

    import sys
    from PIL import Image

    

    
    images=[Image.open(x) for x in list_image[start:end]]
    widths, heights = zip(*(i.size for i in images))

    total_width = sum(widths)
    max_height = max(heights)

    new_im = Image.new('RGB', (total_width, max_height))

    x_offset = 0
    for im in images:
        new_im.paste(im, (x_offset,0))
        x_offset += im.size[0]

    new_im.save('test%s.jpg'%str(i))

def save_as_pdf(pdf_name):
    
    

    for i in range(9):
        merge(i*5,(i+1)*5,i)

    import os
    import img2pdf

    with open(pdf_name, "wb") as f:
        f.write(img2pdf.convert([i for i in os.listdir('/Users/mac/Desktop/pca_knn_cancer_drug/pca_knn/histogram') if i.endswith(".jpg")]))
```
